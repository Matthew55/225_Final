```python
import pickle
import matplotlib.pyplot as plt
import numpy
```

```python
with open('01-1to8.data', 'rb') as filehandle:
    data01 = pickle.load(filehandle)
with open('001-1to8.data', 'rb') as filehandle:
    data001 = pickle.load(filehandle)
with open('0001-1to5.data', 'rb') as filehandle:
    data0001a = pickle.load(filehandle)                          
with open('0001-6to8.data', 'rb') as filehandle:
    data0001b = pickle.load(filehandle)
```

```python
# We trained 24 models and now have to analyze the data
# The data is stored in groups aggreagated by learning rate
# Since we have 24 models total and 3 seperate learning rates used, we will get 3 groups of 8 
# Here we print out  the parameters for each of the models (This is just a copy of the training loop
# removing the actual training code)
bspi = [128, 512]
momen = [0, .9]
warmupiters = [0, 1000]
lr = [.01,.001,.0001]
# lr = [.0001]
for l in lr:
    for w in warmupiters:
        for m in momen:
            for i in bspi:
                print([l,w,m,i])

```

```python
def get_info(data):
    
    ### Each data file contains models with a learning rate equal to the last 4 digits of the parameter name
    ### We train 8 models for each of the 3 different learning rates 
    ### In each of the 8 models we vary the momemtum values, the warm up iterations, the proposal number
    ### this function extracts the final training and validation loss value for each model
    
    ### Below is a list containing the parameters for each of the 4 models where you would
    ### Replace the first element with the respective learning rate
    ### The other 3 elements are the Warm up iterations, Momentum value, Propsoal # respectively 
    
    # [0.0001, 0, 0, 128]
    # [0.0001, 0, 0, 512]
    # [0.0001, 0, 0.9, 128]
    # [0.0001, 0, 0.9, 512]
    # [0.0001, 1000, 0, 128]
    # [0.0001, 1000, 0, 512]
    # [0.0001, 1000, 0.9, 128]
    # [0.0001, 1000, 0.9, 512]

    [iters1,totloss,iters2,valloss] = data
    for i,model in enumerate(totloss):
        finalloss = model[-1]
        print("model" + str(i+1) + " total training loss value = " + str(finalloss))
        
    for i,model in enumerate(valloss):
        finalloss = model[-1]
        print("model" + str(i+1) + " total validation loss value = " + str(finalloss))
    
    
```

```python
# Loss values for the first group of 8 models with learning rate = .01
get_info(data01)
```

```python
# Loss values for the second group of 8 models with learning rate = .001
get_info(data001)
```

```python
# Loss values for the third group 8 models with learning rate = .0001
# the seond data file picks up where the first one leaves off
get_info(data0001a)
print("---------Next 3 models begins here, model 1 is actually model 6 and so on------------")
get_info(data0001b)
```

```python
# Plot the Loss curve of the winning model (With the lowest validation loss)
# This model occurs at model 6 (model 5 in here in the python list with 0 based indexing) 
# in group 2 and has parameters [0.001, 1000, 0, 512]
[iters1,totalloss,iters2,valloss] = data001
plt.plot(iters1[5],totalloss[5])
plt.plot(iters2[5],valloss[5])
plt.legend(['total_loss', 'validation_loss'], loc='upper left')
plt.xlabel("Iteration")
plt.ylabel("Loss")
plt.title("mask_rcnn_R_101_FPN_3x")
plt.show()
print("ultimate validation loss: " + str(valloss[5][-1]))
print("Winning parameter values: " + str("LR = 0.001, Warm up Iterms = 1000, Momemtum = 0, ROI Propsoal 512"))

```
